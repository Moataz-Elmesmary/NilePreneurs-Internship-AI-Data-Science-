```python
import numpy as np
```

## Linear Combination


```python
v1 = np.array([[1],
               [2],
               [3]])

v2 = np.array([[3],
               [5],
               [1]])

v3 = np.array([[0],
               [0],
               [8]])

b = 3 * v1
print(b)
```

```python
x = -v1 + v2 - 0.5 * v3
print(x)
```

## Vector Span


```python
A = np.array([[3, -5, 1],
              [-1, 0, 7],
              [2, 1, -4]])

B = np.array([[19],
              [10],
              [-1]])

np.linalg.inv(A)@B
```

```python
#Using sympy

import sympy as sm
M = sm.Matrix([[3, -5, 1, 19],
               [-1, 0, 7, 10],
               [2, 1, -4, -1]])
M.rref()
```
