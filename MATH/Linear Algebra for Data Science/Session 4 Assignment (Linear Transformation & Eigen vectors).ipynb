## Linear Transformation
### For the following transformation matrix:
<b>$$\begin{bmatrix} 1 & 2 \\ 3 & 7 \end{bmatrix}$$</b>
### Find the result if this matrix applied to the following vectors:
<b>$$\begin{bmatrix} 1 \\ 0 \end{bmatrix}$$</b> 
<b>$$\begin{bmatrix} 0 \\ 1 \end{bmatrix}$$</b>
### If we apply this matrix to a vector (x,y) and the result is a vector (1,0). What are the values of x and y?

```python
A = np.array([[1, 2],
              [3, 7]])

v1 = np.array([1, 0])
v2 = np.array([0, 1])

NewB1 = A@v1
NewB2 = A@v2

print(NewB1, NewB2)
```

```python
np.linalg.inv(A)@v1
```

## Matrix Determinant


```python
a1 = np.array([[3, 8],
               [4, 6]])
a2 = np.array([[4, 6],
               [3, 8]])
a3 = np.array([[6, 1, 1],
               [4, -2, 5],
               [2, 8, 7]])
a4 = np.array([[3, 2, 0, 1],
               [4, 0, 1, 2],
               [3, 0, 2, 1],
               [9, 2, 3, 1]])
a5 = np.array([[1, 2, 4, 4],
               [3, 4, 8, 0]])
a6 = np.array([[1, 2, 3],
               [2, 3, 5],
               [3, 4, 7],
               [5, 4, 9]])

print(np.linalg.det(a1))
print(np.linalg.det(a2))
print(np.linalg.det(a3))
print(np.linalg.det(a4))
#print(np.linalg.det(a5))  not square matrix
#print(np.linalg.det(a6))  not square matrix
```

## Matrix Inverse


```python
print(np.linalg.inv(a1))
print(np.linalg.inv(a2))
print(np.linalg.inv(a3))
print(np.linalg.inv(a4))
#print(np.linalg.inv(a5))  not square matrix
#print(np.linalg.inv(a6))  not square matrix
```

## Eigen Values and Eigen Vectors
### Find eigen values and eigen vectors for the following matrices:
<b>$$\begin{bmatrix} 1 & 0 \\ 0 & 2 \end{bmatrix}$$</b>
<b>$$\begin{bmatrix} 1 & 2 \\ 3 & -4 \end{bmatrix}$$</b>

```python
A = np.array([[1, 0],
              [0, 2]])
B = np.array([[1, 2],
              [3, -4]])
```

```python
np.linalg.eig(A)
```

```python
np.linalg.eig(B)
```
