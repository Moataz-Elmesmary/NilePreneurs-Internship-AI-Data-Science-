```python
import numpy as np
```

## Represent in paper and as numpy arrays Ax=b


```python
A1 = np.array([[3, -1], [1, 0]])
B1 = np.array([[2], [4]])

A2 = np.array([[3, -2], [2, -2]])
B2 = np.array([[7], [2]])

A3 = np.array([[3, -2, 1], [1, 1, 1], [3, -2, -1]])
B3 = np.array([[7], [2], [3]])
```

### Find the solution of the above system of linear equations

```python
sol1 = np.linalg.inv(A1)@B1  #or  sol1 = np.linalg.solve(A1, B1)
sol2 = np.linalg.inv(A2)@B2
sol3 = np.linalg.inv(A3)@B3

print(sol1)
print(sol2)
print(sol3)
```

## Calculate in paper then confirm your results using numpy.


```python
A1 = np.array([[3, 1, 0],
               [2, 1, 2]])
B1 = np.array([[1, 3],
               [2, 5],
               [-1, 3]])

C1 = A1@B1

print("AB = ",C1)
print("Transpose Of C1 = ",C1.T)
```

```python
A2 = np.array([[1, 0, 3],
              [4, 5, -1],
              [0, 0, 2]])
B2 = np.array([[1],
               [2],
               [5]])

C2 = A2@B2

print("A2B2 = ",C2)
print("Transpose Of C2 = ",C2.T)
```
