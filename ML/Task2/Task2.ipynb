```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('car_emission.csv', encoding = 'ISO-8859-1')
df.head(7)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.dropna(how = 'all',inplace = True)
```

```python
df.isnull().sum()
```

```python
df.dropna(how = 'all',inplace = True)
```

```python
df.drop(['Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33', 'Electric energy consumption Miles/kWh','wh/km','Maximum range (Km)','Maximum range (Miles)','Electricity cost'], axis=1, inplace=True)
```

```python
df = df.loc[:, df.isin([' ','NULL',0]).mean() < .6]
df.head()
```

```python
df['Total cost / 12000 miles'] = df['Total cost / 12000 miles'].apply(lambda x : str(x).replace("£",""))
df['Fuel Cost 12000 Miles'] = df['Fuel Cost 12000 Miles'].apply(lambda x : str(x).replace("£",""))

df['Total cost / 12000 miles'] = df['Total cost / 12000 miles'].str.replace(",","").astype(float)
df['Fuel Cost 12000 Miles'] = df['Fuel Cost 12000 Miles'].str.replace(",","").astype(float)
```

```python
df['Emissions NOx [mg/km]'].fillna(df['Emissions NOx [mg/km]'].mean(),inplace=True)
df['Imperial Combined'].fillna(df['Imperial Combined'].mean(),inplace=True)
df['Metric Urban (Cold)'].fillna(df['Metric Urban (Cold)'].mean(),inplace=True)
df['Metric Extra-Urban'].fillna(df['Metric Extra-Urban'].mean(),inplace=True)
df['Engine Capacity'].fillna(df['Engine Capacity'].mean(),inplace=True)
df['Metric Combined'].fillna(df['Metric Combined'].mean(),inplace=True)
df['Imperial Urban (Cold)'].fillna(df['Imperial Urban (Cold)'].mean(),inplace=True)
df['Imperial Extra-Urban'].fillna(df['Imperial Extra-Urban'].mean(),inplace=True)
df['Fuel Cost 12000 Miles'].fillna(df['Fuel Cost 12000 Miles'].mean(),inplace=True)
df['Emissions CO [mg/km]'].fillna(df['Emissions CO [mg/km]'].mean(),inplace=True)
df['THC Emissions [mg/km]'].fillna(df['THC Emissions [mg/km]'].mean(),inplace=True)
df['THC + NOx Emissions [mg/km]'].fillna(df['THC + NOx Emissions [mg/km]'].mean(),inplace=True)
df['Particulates [No.] [mg/km]'].fillna(df['Particulates [No.] [mg/km]'].mean(),inplace=True)
df['Transmission'].fillna(method='bfill', inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
for i in ['Transmission','Fuel Type']:
  df[i]=encode.fit_transform(df[i])
```

```python
df.head()
```

```python
X = df[['Transmission', 'Engine Capacity', 'Fuel Type', 'Metric Combined', 'Imperial Combined']]  
y = df['Emissions CO [mg/km]']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```
